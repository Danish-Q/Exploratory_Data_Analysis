#Which hosts are the busiest and why?
#Most of the busiest host are located in Manhattan and Most busiest host's neighbourhood is financial district, So maybe the host is the busiest and rents his place most of the time as compared to others. Plus, other evidence is the price which is average as compared to others in the same area.

```python
AB_1 = ps.sqldf("""

SELECT *, COUNT(*) as frequency
FROM AB_NYC
GROUP BY host_id
ORDER BY frequency desc


""")
AB_1
```

#What areas have more traffic than others, and why is that the case?
#Traffic is mostly at the places which have low prices and are near the major cities.

```python
AB_2 = ps.sqldf("""

SELECT *,COUNT(latitude) as traffic
FROM AB_NYC
GROUP BY neighbourhood
ORDER BY traffic desc

""")
AB_2
```

#Are there any relationships between prices, number of reviews, and the number of days that a given listing is booked?
#There is no relationship in price, number of reviews and number of nights as per the data if the price is high the number of review and number of nights is low, and if the number of reviews is high then the price is low and number of nights is also low.

```python
AB_3 = ps.sqldf("""

SELECT price, number_of_reviews, minimum_nights
FROM AB_NYC
ORDER BY minimum_nights desc, number_of_reviews desc, price desc

""")
AB_3
```
